```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/train.csv")
df.head()
```

```python
df.columns
```

```python
filt = df['Arrival Delay in Minutes'].isnull()
```

```python
sati = df[filt].groupby(['Class', 'satisfaction'], as_index=False).size().pivot_table(index='Class', columns='satisfaction')
```

```python
filt = sati['size']['neutral or dissatisfied'] < sati['size']['satisfied']
```
