```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv', index_col='Location')
```

```python
df
```

```python
df1 = df.iloc[:4, :]
df2 = df.iloc[4:, :]
```

```python
tt = pd.concat([df1, df2])
```

```python
tt
```

```python
df3 = df.iloc[:2,:4]
df4 = df.iloc[5:,3:]
```

```python
tt = pd.concat([df3, df4], join='inner')
```

```python
tt
```

```python
tt = pd.concat([df3, df4]).fillna(0)
tt
```

```python
df5 = df.T.iloc[:7,:3]
df6 = df.T.iloc[6:,2:5]

display(df5)
display(df6)
```

```python
tt = pd.merge(df5, df6, on='Algeria')
tt
```

```python
tt = pd.merge(df5, df6, on='Algeria', how='outer')
tt
```
