```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/timeTest.csv')
```

```python
df.info()
```

```python
df['Yr_Mo_Dy'] = pd.to_datetime(df['Yr_Mo_Dy'])
df['Yr_Mo_Dy']
```

```python
df['Yr_Mo_Dy'].dt.year.unique()
```

```python
def fix_year(x):
    import datetime
    
    year = x.year - 100 if x.year >= 2061 else x.year
    return pd.to_datetime(datetime.date(year, x.month, x.day))

df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(fix_year)
df['Yr_Mo_Dy']
```

```python
temp = df.groupby(df['Yr_Mo_Dy'].dt.year).mean()
temp
```

```python
df['weekday'] = df['Yr_Mo_Dy'].dt.weekday
df['weekday']
```

```python
df['WeekCheck'] = df['weekday'].apply(lambda x:1 if x >= 5 else 0)
df['WeekCheck']
```

```python
df.groupby(df['Yr_Mo_Dy'].dt.month).mean()
```

```python
df = df.fillna(method='ffill').fillna(method='bfill')
df.isnull().sum()
```

```python
df.groupby(df['Yr_Mo_Dy'].dt.to_period('M')).mean()
```
