```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv')
```

```python
df.head(5)
```

```python
df.groupby('host_name').size().head(5)
```

```python
df.groupby('host_name').size().to_frame().rename(columns={0:'counts'}).sort_values('counts', ascending=False)
```

```python
df.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size()
```

```python
df.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size().groupby('neighbourhood_group', as_index=False).max()
```

```python
df[['neighbourhood_group', 'price']].groupby('neighbourhood_group').agg(['mean', 'var', 'max', 'min'])
```

```python
df[['neighbourhood_group', 'reviews_per_month']].groupby('neighbourhood_group').agg(['mean', 'var', 'max', 'min'])
```

```python
df[['neighbourhood', 'neighbourhood_group', 'price']].groupby(['neighbourhood', 'neighbourhood_group']).agg(['mean'])
```

```python
df.groupby(['neighbourhood', 'neighbourhood_group'])['price'].agg(['mean']).unstack()
```

```python
df.groupby(['neighbourhood', 'neighbourhood_group'])['price'].agg(['mean']).unstack().fillna(-999)
```

```python
filt = df['neighbourhood_group'] == 'Queens'
df[filt].groupby('neighbourhood')['price'].agg(['mean', 'var', 'max', 'min']).head(5)
```
