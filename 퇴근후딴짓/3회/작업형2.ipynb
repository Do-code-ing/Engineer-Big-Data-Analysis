```python
import pandas as pd

train = pd.read_csv('t2-1-train.csv')
test = pd.read_csv('t2-1-test.csv')
```

```python
train.head(3)
```

```python
test.head(3)
```

```python
train.AnnualIncome = train.AnnualIncome.fillna(train.AnnualIncome.mean())
```

```python
y_train = train.pop('TravelInsurance')
```

```python
drop_cols = ['id']
train = train.drop(columns=drop_cols)
test = test.drop(columns=drop_cols)
```

```python
train = pd.get_dummies(train)
test = pd.get_dummies(test)
```

```python
test.columns.difference(train.columns)
```

```python
train = train.reindex(columns=train.columns.tolist()+['Employment Type_Casual employment'], fill_value=0)
```

```python
train = train.sort_index(axis=1)
test = test.sort_index(axis=1)
```

```python
from sklearn.model_selection import train_test_split

x_tr, x_val, y_tr, y_val = train_test_split(train, y_train, test_size=0.3, random_state=0, stratify=y_train)
```

```python
x_tr
```

```python
test.reindex()
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
x_tr = scaler.fit_transform(x_tr)
x_val = scaler.transform(x_val)
test = scaler.transform(test)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=0, max_depth=7, n_estimators=100)
model.fit(x_tr, y_tr)
```

```python
pred_proba = model.predict_proba(x_val)[:, 1]
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_val, pred_proba)
```
